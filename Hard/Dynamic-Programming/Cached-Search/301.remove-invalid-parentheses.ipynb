## Attempt 1 - Accepted

```python
class Solution:
    def removeInvalidParentheses(self, s: str) -> List[str]:
        curr = [("", 0)]
        for ch in s:
            nexxt = []
            if ch not in {"(", ")"}:
                for c in curr:
                    nexxt.append((c[0] + ch, c[1]))
                curr = nexxt
                continue
            
            if ch == "(":
                for c in curr:
                    nexxt.append((c[0] + ch, c[1] + 1))
            else:
                for c in curr:
                    if c[1] <= 0:
                        continue
                    nexxt.append((c[0] + ch, c[1] - 1))
            
            curr += nexxt
        
        maxLen = max([len(c[0]) for c in curr if c[1] == 0])
        res = list(set([c[0] for c in curr if len(c[0]) == maxLen and c[1] == 0]))
        return res
```

```python
class Solution:
    def removeInvalidParentheses(self, s: str) -> List[str]:
        record = [("", 0, 0)]
        for c in s:
            new_record = []
            for r in record:
                sub, remove_cnt, bal = r
                if c == "(":
                    new_record.append(tuple([sub + c, remove_cnt, bal + 1]))
                    new_record.append(tuple([sub, remove_cnt + 1, bal]))
                elif c == ")":
                    new_record.append(tuple([sub, remove_cnt + 1, bal]))
                    if bal > 0:
                        new_record.append(tuple([sub + c, remove_cnt, bal - 1]))
                else:
                    new_record.append(tuple([sub + c, remove_cnt, bal]))
            record = new_record
        
        record = [r for r in record if r[2] == 0]
        record.sort(key=lambda x: x[1])
        minn_rm = record[0][1]
        record = [r for r in record if r[1] == minn_rm]
        return list(set([r[0] for r in record]))
```

## Attempt 4 - Accepted

```python
class Solution:
    def removeInvalidParentheses(self, s: str) -> List[str]:
        res = set([("", 0)])
        for c in s:
            new_res = {r for r in res}
            if c == '(':
                for r, cnt in res:
                    new_res.add((r + '(', cnt + 1))
            elif c == ')':
                for r, cnt in res:
                    if cnt > 0:
                        new_res.add((r + ')', cnt - 1))
            else:
                for r, cnt in res:
                    new_res.add((r + c, cnt))
            
            res = new_res
            # print(res)
        
        res = {r for r, cnt in res if cnt == 0}
        if len(res) < 1:
            return res
        max_len = max([len(r) for r in res])
        res = list([r for r in res if len(r) == max_len])
        return res
```
