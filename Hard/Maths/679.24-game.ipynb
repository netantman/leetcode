```python
class Solution:
    def judgePoint24(self, cards: List[int]) -> bool:
        tol = 1e-6
        self.record = dict()
        
        def possibles(a, b):
            res = set([float(a + b), float(a - b), float(b - a), float(a * b)])
            if b != 0:
                res.add(a / b)
            if a != 0:
                res.add(b / a)
            return res
        
        def check(cards):
            cards = sorted(cards)
            if tuple(cards) in self.record:
                return self.record[tuple(cards)]
            
            if len(cards) == 2:
                a, b = tuple(cards)
                res = False
                for n in possibles(a, b):
                    if abs(n - 24.0) < tol:
                        res = True
                        break
                self.record[tuple(cards)] = res
                return res
            if len(cards) == 3:
                a, b, c = tuple(cards)
                ab = possibles(a, b)
                for n in ab:
                    if check([n, c]):
                        self.record[tuple(cards)] = True
                        return True
                ac = possibles(a, c)
                for n in ac:
                    if check([n, b]):
                        self.record[tuple(cards)] = True
                        return True
                bc = possibles(b, c)
                for n in bc:
                    if check([n, a]):
                        self.record[tuple(cards)] = True
                        return True
                self.record[tuple(cards)] = False
                return False
            
            if len(cards) == 4:
                a, b, c, d = tuple(cards)
                ab = possibles(a, b)
                for n in ab:
                    if check([n, c, d]):
                        # self.record[tuple(cards)] = True
                        return True
                ac = possibles(a, c)
                for n in ac:
                    if check([n, b, d]):
                        # self.record[tuple(cards)] = True
                        return True
                ad = possibles(a, d)
                for n in ad:
                    if check([n, b, c]):
                        # self.record[tuple(cards)] = True
                        return True
                bc = possibles(b, c)
                for n in bc:
                    if check([n, a, d]):
                        # self.record[tuple(cards)] = True
                        return True
                bd = possibles(b, d)
                for n in bd:
                    if check([n, a, c]):
                        # self.record[tuple(cards)] = True
                        return True
                cd = possibles(c, d)
                for n in cd:
                    if check([n, a, b]):
                        # self.record[tuple(cards)] = True
                        return True
                return False
        
        res = check(cards)
        # print(self.record)
        # print(8 / (3 - 8/3))
        return res
            
```

```python
class Solution {
public:
    bool inner(vector<float>& cards) {
        float tol = 0.00001;
        int L = cards.size();
        if (L == 1) {
            bool res = std::abs(cards[0] - 24.0) < tol;
        }
        if (L == 2) {
            float a = cards[0];
            float b = cards[1];
            if (std::abs( a + b - 24.) < tol) {
                return true;
            }
            if (std::abs( a - b - 24.) < tol) {
                return true;
            }
            if (std::abs( b - a - 24.) < tol) {
                return true;
            }
            if (std::abs( a * b - 24.) < tol) {
                return true;
            }
            if ((std::abs(a) > tol) && std::abs(b / a - 24.) < tol) {
                return true;
            }
            if ((std::abs(b) > tol) && std::abs(a / b - 24.) < tol) {
                return true;
            }
            return false;
        }
        
        for (int idx = 0; idx < L; idx++) {
            for (int jdx = idx + 1; jdx < L; jdx++) {
                vector<float> temp = {};
                float a = cards[idx];
                float b = cards[jdx];
                temp.push_back(a + b);
                temp.push_back(a - b);
                temp.push_back(b - a);
                temp.push_back(a * b);
                if (std::abs(a) > tol) {
                    temp.push_back(b / a);
                }
                if (std::abs(b) > tol) {
                    temp.push_back(a / b);
                }
                
                vector<float> new_cards = {};
                for (int kdx = 0; kdx < L; kdx++) {
                    if ((kdx == idx) || (kdx == jdx)) {
                        continue;
                    }
                    new_cards.push_back(cards[kdx]);
                }
                for (float t : temp) {
                    vector<float> combined = new_cards;
                    combined.push_back(t);
                    if (inner(combined)) {
                        return true;
                    }
                }
            }
        }
        return false;
    }
    
    bool judgePoint24(vector<int>& cards) {
        vector<float> float_cards = {};
        for (int c : cards) {
            float_cards.push_back(float(c));
        }
        return inner(float_cards);
    }
};
```
