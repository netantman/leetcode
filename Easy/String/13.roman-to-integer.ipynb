```python
class Solution:
    def romanToInt(self, s: str) -> int:
        if s=="":
            return 0
        
        res=0
        leading=0
        s_len=len(s)
        while leading<s_len:
            if s[leading]=="M":
                leading+=1
                res+=1000
            else:
                break
        s=s[leading:]
        
        if s=="":
            return res
        if s[:2]=="CM":
            res+=900
            s=s[2:]
        elif s[:2]=="CD":
            res+=400
            s=s[2:]
        else: 
            if s[:1]=="D":
                res+=500
                s=s[1:]
            leading=0
            s_len=len(s)
            while leading<s_len:
                if s[leading]=="C":
                    leading+=1
                    res+=100
                else:
                    break
            s=s[leading:]
        
        if s=="":
            return res
        if s[:2]=="XC":
            res+=90
            s=s[2:]
        elif s[:2]=="XL":
            res+=40
            s=s[2:]
        else:
            if s[:1]=="L":
                res+=50
                s=s[1:]
            leading=0
            s_len=len(s)
            while leading<s_len:
                if s[leading]=="X":
                    leading+=1
                    res+=10
                else:
                    break
            s=s[leading:]
        
        if s=="":
            return res
        if s[:2]=="IX":
            res+=9
            s=s[2:]
        elif s[:2]=="IV":
            res+=4
            s=s[2:]
        else: 
            if s[:1]=="V":
                res+=5
                s=s[1:]
            leading=0
            s_len=len(s)
            while leading<s_len:
                if s[leading]=="I":
                    leading+=1
                    res+=1
                else:
                    break
        
        return res
```
