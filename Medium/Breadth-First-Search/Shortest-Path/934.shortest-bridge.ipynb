```python
class Solution:
    def __init__(self):
        self.R=None
        self.C=None
    
    def shortestBridge(self, A: List[List[int]]) -> int:
        self.R,self.C=len(A),len(A[0])
        def get_neighbors(r,c):
            res=[]
            if r-1>=0:
                res.append((r-1,c))
            if r+1<self.R:
                res.append((r+1,c))
            if c-1>=0:
                res.append((r,c-1))
            if c+1<self.C:
                res.append((r,c+1))
            return res
        q=[]
        found=False
        for r in range(self.R):
            for c in range(self.C):
                if A[r][c]>0:
                    q.append((r,c))
                    found=True
                    break
            if found:
                break
        
        res=[[float("inf") for c in range(self.C)] for r in range(self.R)] 
        res[q[0][0]][q[0][1]]=0
        qq=[]
        while len(q)>0:
            head=q.pop(0)
            qq.append(head)
            neighbors=get_neighbors(*head)
            for r,c in neighbors:
                if A[r][c]>0 and res[r][c]>0:
                    res[r][c]=0
                    q.append((r,c))
                    
        q=[elem for elem in qq]
        while len(q)>0:
            head=q.pop(0)
            lvl=res[head[0]][head[1]]
            neighbors=get_neighbors(*head)
            for r,c in neighbors:
                if res[r][c]>lvl+1:
                    res[r][c]=lvl+1
                    if A[r][c]>0:
                        return res[r][c]-1
                    else:
                        q.append((r,c))
```

## Attempt 3 - Accepted

```python
class Solution:
    def shortestBridge(self, grid: List[List[int]]) -> int:
        N = len(grid)
        island = set()
        q = collections.deque([])

        for r in range(N):
            for c in range(N):
                if grid[r][c] == 0:
                    continue
                
                island.add((r, c))
                q.append((r, c))
                # print(island)
                while q:
                    rr, cc = q.popleft()
                    # print(rr, cc)
                    if rr - 1 >= 0 and grid[rr - 1][cc] == 1 and (rr - 1, cc) not in island:
                        island.add((rr - 1, cc))
                        q.append((rr - 1, cc))
                    if rr + 1 < N and grid[rr + 1][cc] == 1 and (rr + 1, cc) not in island:
                        island.add((rr + 1, cc))
                        q.append((rr + 1, cc))
                    if cc - 1 >= 0 and grid[rr][cc - 1] == 1 and (rr, cc - 1) not in island:
                        island.add((rr, cc - 1))
                        q.append((rr, cc - 1))
                    if cc + 1 < N and grid[rr][cc + 1] == 1 and (rr, cc + 1) not in island:
                        island.add((rr, cc + 1))
                        q.append((rr, cc + 1))

                break
            if len(island) > 0: # this tripped you: when an island is found, this needs to break too.
                break
        
        for r, c in island:
            grid[r][c] = 0

        
        # print(grid)
        
        q = collections.deque(island)
        sp = {isl: 0 for isl in island}
        while q:
            rr, cc = q.popleft()
            if rr - 1 >= 0 and sp.get((rr - 1, cc), float('inf')) > sp[(rr, cc)] + 1:
                sp[(rr - 1, cc)] = sp[(rr, cc)] + 1
                if grid[rr - 1][cc] == 1:
                    return sp[(rr - 1, cc)] - 1
                q.append((rr - 1, cc))
            if rr + 1 < N and sp.get((rr + 1, cc), float('inf')) > sp[(rr, cc)] + 1:
                sp[(rr + 1, cc)] = sp[(rr, cc)] + 1
                if grid[rr + 1][cc] == 1:
                    return sp[(rr + 1, cc)] - 1
                q.append((rr + 1, cc))
            if cc - 1 >= 0 and sp.get((rr, cc - 1), float('inf')) > sp[(rr, cc)] + 1:
                sp[(rr, cc - 1)] = sp[(rr, cc)] + 1
                if grid[rr][cc - 1] == 1:
                    return sp[(rr, cc - 1)] - 1
                q.append((rr, cc - 1))
            if cc + 1 < N and sp.get((rr, cc + 1), float('inf')) > sp[(rr, cc)] + 1:
                sp[(rr, cc + 1)] = sp[(rr, cc)] + 1
                if grid[rr][cc + 1] == 1:
                    return sp[(rr, cc + 1)] - 1
                q.append((rr, cc + 1))

        # should not reach here
        return N
```

## Lessons and Thoughts
- BFS from multiple points to multiple points. When doing these shortest path problems, probably just maintain the record of shortest paths so far and a queue is the easiest to do.
- Note that `l=s[:]` is not deep copy of a list!
- Attempt 3 seems cleaner, although it is the same multi-point to multi-point algo. The way is to just find one island, and use that island to find another. Note the optimization still hold that once a point on the other side is found, we can early stop.
