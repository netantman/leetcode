## Attempt 2 - Accepted

```python
class Solution:
    def search(self, nums: List[int], target: int) -> int:
        if len(nums)<1:
            return -1
        if len(nums)==1:
            if nums[0]==target:
                return 0
            else:
                return -1
        
        len_nums=len(nums)
        left, right=0, len_nums-1
        if (nums[left]<nums[right]):
            if (target<nums[left]) or (target>nums[right]):
                return -1
        
        while left<right-1:
            mid=(left+right)//2
            if nums[left]<nums[right]:
                if nums[mid]<target:
                    left=mid
                elif nums[mid]>target:
                    right=mid
                else:
                    return mid
            else: 
                if nums[mid]>nums[left]:
                    if nums[mid]==target:
                        return mid
                    elif nums[mid]<target:
                        left=mid
                    else: # nums[mid]>target
                        if nums[left]==target:
                            return left
                        elif nums[left]<target:
                            right=mid
                        else:
                            left=mid
                else: # nums[mid]<nums[left]
                    if nums[mid]==target:
                        return mid
                    elif nums[mid]>target:
                        right=mid
                    else: # nums[mid]<target
                        if nums[right]==target:
                            return right
                        elif nums[right]<target:
                            right=mid
                        else:
                            left=mid
                            
        if nums[left]==target:
            return left
        if nums[right]==target:
            return right
        
        return -1
```
