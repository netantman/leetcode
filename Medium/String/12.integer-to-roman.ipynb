## Attempt 3 - Accepted

```python
class Solution:
    def intToRoman(self, num: int) -> str:
        num = str(num)
        while len(num) < 4:
            num = "0" + num
        
        thousand, hundred, ten, o = tuple([c for c in num])
        thousand, hundred, ten, o = int(thousand), int(hundred), int(ten), int(o)
        res = ""
        
        for _ in range(thousand):
            res += "M"
        
        def add_digit(digit, five, ten, one):
            res = ""
            if digit == 9:
                res += (one + ten)
            elif digit >= 5:
                res += five
                inc = digit - 5
                for _ in range(inc):
                    res += one
            elif digit == 4:
                res += (one + five)
            else:
                for _ in range(digit):
                    res += one
            return res
    
        res += add_digit(hundred, "D", "M", "C")
        res += add_digit(ten, "L", "C", "X")
        res += add_digit(o, "V", "X", "I")
        
        return res
```
