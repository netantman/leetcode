## Attempt 9 - Accepted

```python
class Solution:
    def hasValidPath(self, grid: List[List[int]]) -> bool:
        directions={1:[(0,1), (0,-1)], 2:[(1,0), (-1,0)], 3:[(1,0), (0,-1)], 4: [(1,0), (0,1)], 5:[(-1,0),(0,-1)], 6:[(-1,0),(0,1)]}
        R,C=len(grid), len(grid[0])
        flag=[[0 for _ in range(C)] for _ in range(R)]
        def dfs(r,c):
            flag[r][c]=1
            if r==R-1 and c==C-1:
                return True
            for x,y in directions[grid[r][c]]:
                if 0<=r+x<R and 0<=c+y<C and flag[r+x][c+y]==0 and (-x,-y) in directions[grid[r+x][c+y]]:
                    if dfs(r+x, c+y):
                        return True
            return False
        return dfs(0,0)
```
