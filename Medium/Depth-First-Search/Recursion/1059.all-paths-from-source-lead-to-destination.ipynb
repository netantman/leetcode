## Attempt 1 - Accepted

```python
class Solution:
    def leadsToDestination(self, n: int, edges: List[List[int]], source: int, destination: int) -> bool:
        record = defaultdict(list)
        for edge in edges:
            a, b = tuple(edge)
            record[a].append(b)
            
        if destination in record:
            return False
        
        dfs = [False for idx in range(n)]
        dfs[destination] = True
        
        paths = [tuple([source])]
        
        while paths:
            top = paths[-1]
            last = top[-1]
            if last not in record:
                if last != destination:
                    return False
                paths.pop()
                continue
            
            for n in record[last]:
                if n in top:
                    return False
                if not dfs[n]:
                    paths.append(tuple(list(top) + [n]))
                    break
            else:
                dfs[last] = True
                paths.pop()
        
        return True
                    
                
```

## Attempt 2 - Accepted, seems essential to pass the time limit that stack implementation of DFS 

```python
class Solution {
public:
    bool leadsToDestination(int n, vector<vector<int>>& edges, int source, int destination) {
        std::map<int, std::vector<int>> record {};
        
        for (std::vector<int> edge : edges) {
            int a = edge.front();
            int b = edge.back();
            if (record.find(a) != record.end()) {
                record[a].push_back(b);    
            }
            else {
                std::vector<int> tmp {b};
                record[a] = tmp;
            }
        }
        
        if (record.find(destination) != record.end()) {
            return false;
        }
        
        std::vector<bool> dfs(n, false);
        dfs[destination] = true;
        
        std::deque<std::vector<int>> paths {};
        std::vector<int> tmp {source};
        paths.push_back(tmp);
        
        while (paths.size() > 0) {
            std::vector<int> top = paths.back();
            int last = top.back();
            
            if (record.find(last) == record.end()) {
                if (last != destination) {
                    return false;
                }
            paths.pop_back();
            continue;
            }
            
            bool flag = false;
            for (int child : record[last]) {
                if (std::find(top.begin(), top.end(), child) != top.end()) { // the way to see if an element is in a vector
                    return false;
                }
                if (dfs[child]) {
                    continue;
                }
                std::vector tmp = top;
                tmp.push_back(child);
                paths.push_back(tmp);
                flag = true;
                break;
            }
            if (not flag) {
                dfs[last] = true;
                paths.pop_back();
            }
        }
        return true;
    }
};
```

## Lessons and Thoughts
- DFS implemented using a stack, rather than a recursion.
- The missed edge case is when `destination` has out-degree.
- Interestingly, it is considered DFS is better than BFS in finding cycles in a graph; see this [stackoverflow](https://stackoverflow.com/questions/2869647/why-dfs-and-not-bfs-for-finding-cycle-in-graphs).
