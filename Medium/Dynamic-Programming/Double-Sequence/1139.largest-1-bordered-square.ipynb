```python
class Solution:
    def largest1BorderedSquare(self, grid: List[List[int]]) -> int:
        R,C=len(grid), len(grid[0])
        up=[[0 for c in range(C)] for r in range(R)]
        down=[[0 for c in range(C)] for r in range(R)]
        left=[[0 for c in range(C)] for r in range(R)]
        right=[[0 for c in range(C)] for r in range(R)]
        
        for r in range(R):
            for c in range(C):
                if grid[r][c]>0:
                    if r==0:
                        up[r][c]=1
                    else:
                        up[r][c]=up[r-1][c]+1
                else:
                    up[r][c]=0
        
        for r in reversed(range(R)):
            for c in range(C):
                if grid[r][c]>0:
                    if r==R-1:
                        down[r][c]=1
                    else:
                        down[r][c]=down[r+1][c]+1
                else:
                    down[r][c]=0
        
        for c in range(C):
            for r in range(R):
                if grid[r][c]>0:
                    if c==0:
                        left[r][c]=1
                    else:
                        left[r][c]=left[r][c-1]+1
                else:
                    left[r][c]=0
        
        for c in reversed(range(C)):
            for r in range(R):
                if grid[r][c]>0:
                    if c==C-1:
                        right[r][c]=1
                    else:
                        right[r][c]=right[r][c+1]+1
                else:
                    right[r][c]=0
                    
        # print(f"left: {left}")
        # print(f"up: {up}")
        # print(f"right: {right}")
        # print(f"down: {down}")
        
        res=0
        for r in range(R):
            for c in range(C):
                far=min(up[r][c], left[r][c])
                side=far
                for gap in range(far-1,-1,-1):
                    side=min(far, down[r-gap][c-gap], right[r-gap][c-gap])
                    # if side>=5:
                        # print(f"r,c:{r,c}")
                        # print(f"gap: {gap}")
                        # print(f"right: {right[r-gap][c-gap]}")
                        # print(f"down: {down[r-gap][c-gap]}")
                    if side>=gap+1:
                        res=max(res, gap+1)
        return res**2
```
