## Attempt 3 - Accepted

```python
class Solution:
    def numSubarraysWithSum(self, nums: List[int], goal: int) -> int:
        cumsum = [0]
        for n in nums:
            cumsum.append(cumsum[-1] + n)

        # print(cumsum)
        
        L = len(nums)
        r, new_left = 0, 0
        cnt = 0
        if goal == 0:
            for l in range(L):
                if l < new_left:
                    continue
                if r == l:
                    r += 1
                
                while r <= L and cumsum[r] == cumsum[l]:
                    r += 1
                cnt += (r - l) * (r - l - 1) // 2
                new_left = r
            return cnt

        for l in range(L):
            if l < new_left:
                continue
            if r == l:
                r += 1
            if new_left < l:
                new_left = l
            
            while r <= L and cumsum[r] - cumsum[l] < goal:
                r += 1
            
            if r > L:
                break
            
            if cumsum[r] - cumsum[l] == goal:
                while new_left < r and cumsum[new_left] == cumsum[l]:
                    new_left += 1
                new_right = r
                while new_right <= L and cumsum[new_right] == cumsum[r]:
                    new_right += 1
                cnt += (new_left - l) * (new_right - r)
        
        return cnt
```
