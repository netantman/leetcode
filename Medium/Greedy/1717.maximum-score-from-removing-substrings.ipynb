## Attempt 3 - Accepted

```python
class Solution:
    def maximumGain(self, s: str, x: int, y: int) -> int:
        def tally(s):
            stack, res = [], 0
            if x >= y:
                for c in s:
                    if c == "b" and stack[-1:] == ["a"]:
                        stack.pop()
                        res += x
                        continue
                    stack.append(c)
                stack2 = []
                while stack:
                    c = stack.pop(0)
                    if c == "a" and stack2[-1:] == ["b"]:
                        stack2.pop()
                        res += y
                        continue
                    stack2.append(c)
            else: # x < y
                for c in s:
                    if c == "a" and stack[-1:] == ["b"]:
                        stack.pop()
                        res += y
                        continue
                    stack.append(c)
                stack2 = []
                while stack:
                    c = stack.pop(0)
                    if c == "b" and stack2[-1:] == ["a"]:
                        stack2.pop()
                        res += x
                        continue
                    stack2.append(c)
            return res
        
        curr = ""
        collect = []
        for c in s:
            if c in {"a", "b"}:
                curr += c
            else:
                collect.append(curr)
                curr = ""
        collect.append(curr)
        collect = [c for c in collect if len(c) > 0]
        return sum(map(tally, collect))
```
